```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_res(csv):
    dqn_single = pd.read_csv(csv, usecols=['Episodic Reward'])
    dqn_single['X'] = range(0,2000)
    sing = dqn_single[dqn_single.X % 10 == 0]
    plt.plot(sing.X, sing['Episodic Reward'], label='Episodic Reward')
    sing['rolling_avg'] = sing['Episodic Reward'].rolling(5).mean()
    plt.plot(sing.X, sing['rolling_avg'], label='Average Episodic Reward')
    plt.legend()
    plt.xlabel('Episodes')
    plt.ylabel('Reward')
```

```python
plot_res('data/Pong-ram-v0_DQN-FC.csv')
```

```python
plot_res('data/Pong-ram-v0_DQN-Single.csv')
```

```python
plot_res('data/Pong-ram-v0_DQN-Single-Equal-Neuron.csv')
```

```python
plot_res('data/Pong-ram-v0_DQN-CNN.csv')
```
